Q1). Write a lambda function that takes an integer 'year' and returns 'age'.

```python
from datetime import datetime

current_year = datetime.now().year
```

```python
current_year = 2026
age_from_year = lambda year: current_year - year

print(age_from_year(1990))
```

Q2) Write a lambda function that takes a float "cost" and returns a float "price" by adding 10% to the cost.

```python
price_from_cost = lambda cost: cost * 1.10
print(price_from_cost(100))

```

Q3) Write a lambda function that takes two strings 'name' , 'domain' and returns an email address (string).

```python
create_email = lambda name, domain: f"{name.strip().lower().replace(' ', '.')}@{domain.strip().lower()}"
print(create_email("John Doe", "example.com"))
```
